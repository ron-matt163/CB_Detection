```python
import re
import string
import stopwords
import nltk # Natural Language tool kit 
nltk.download('stopwords')
from nltk.corpus import stopwords
def message_cleaning(message):
    message = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",message).split())
    Test_punc_removed = [char for char in message if char not in string.punctuation]
    Test_punc_removed_join = ''.join(Test_punc_removed)
    Test_punc_removed_join_clean = [word.lower() for word in Test_punc_removed_join.split() if word.lower() not in stopwords.words('english')]
    print(Test_punc_removed_join_clean)
```

```python
count = 0
for value in trends: 
    for trend in value['trends']:
        if isEnglish(trend['name']): 
            print(trend['name'])
            tweets = tweepy.Cursor(api.search, q=trend['name'],lang="en", since=date_since,result_type='popular', timeout=999999).items(10)
            for tweet in tweets:
                print("Tweet:\n")
                message_cleaning(tweet.text)
                tweet_id = tweet.id
                print("Tweet ID: " + str(tweet_id))
                poster = tweet.user.screen_name
                print("\nPoster: " + poster + "\n")
                print("Replies to this tweet: \n")
                replies = tweepy.Cursor(api.search,q="@"+poster, since_id=tweet_id).items(1000)
                #print(type(replies))
                count = 0
                for reply in replies:
                    if(reply.in_reply_to_status_id == tweet_id):
                        message_cleaning(reply.text)
                        count +=1
                    if count==10:
                        break                
            count +=1
        if count==5:
            break
                
```
