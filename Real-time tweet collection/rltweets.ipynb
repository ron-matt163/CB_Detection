# EXTRACTING TWEETS FROM TRENDING TOPICS

```python
import tweepy
import os
import json
import sys
import geocoder
from langdetect import detect

def isEnglish(s):
    try:
        s.encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        return False
    else:
        return True

# API Keys and Tokens

# Authorization and Authentication

date_since="2020-12-01"

if __name__ == "__main__":
    # Available Locations
    available_loc = api.trends_available()
    # writing a JSON file that has the available trends around the world
    with open("available_locs_for_trend.json","w") as wp:
        wp.write(json.dumps(available_loc, indent=1))

    # Trends for Specific Country     # location as argument variable 
    g = geocoder.osm('India') # getting object that has location's latitude and longitude

    closest_loc = api.trends_closest(g.lat, g.lng)
    trends = api.trends_place(closest_loc[0]['woeid'],count=200,lang="en")
    # writing a JSON file that has the latest trends for that location
    print("The top trends for the location are :") 
    for value in trends: 
        for trend in value['trends']:
            if isEnglish(trend['name']):
                print(trend['name'])                                 
```

```python
count = 0
for value in trends: 
    for trend in value['trends']:
        if isEnglish(trend['name']): 
            print(trend['name'])
            tweets = tweepy.Cursor(api.search, q=trend['name'],lang="en", since=date_since,result_type='popular').items(20)
            for tweet in tweets:
                print("Tweet:\n")
                print(tweet.text)
                tweet_id = tweet.id
                print("Tweet ID: " + str(tweet_id))
                poster = tweet.user.screen_name
                print("\nPoster: " + poster + "\n")
                print("Replies to this tweet: \n")
                replies = tweepy.Cursor(api.search,q="@"+poster, since_id=tweet_id).items()
                #print(type(replies))
                count = 0
                for reply in replies:
                    if(reply.in_reply_to_status_id == tweet_id):
                        print(reply.text)
                        count +=1
                    if count==50:
                        break                
            count +=1
        if count==5:
            break
                
                
```

# Streaming replies of tweets

```python
name="anishgiri"

replies=[] 
non_bmp_map = dict.fromkeys(range(0x10000, sys.maxunicode + 1), 0xfffd)  
for full_tweets in tweepy.Cursor(api.user_timeline,screen_name=name,timeout=999999).items(10):
  for tweet in tweepy.Cursor(api.search,q='to:'+name,result_type='recent',timeout=999999).items(1000):
    if hasattr(tweet, 'in_reply_to_status_id_str'):
      if (tweet.in_reply_to_status_id_str==full_tweets.id_str):
        replies.append(tweet.text)
  print("Tweet :",full_tweets.text.translate(non_bmp_map))
  for elements in replies:
       print("Replies :",elements)
  replies.clear()
```
